```python
f = open('keys/.openai_api_key.txt')
OPENAI_API_KEY = f.read()
client = OpenAI(api_key=OPENAI_API_KEY)
```

```python
def chat_bot(prompt):
    response = client.chat.completions.create(model='gpt-4o-mini', 
                                              messages=[{"role":"system",
                                                         "content":"""You are a professional Education counsellor working with a Data Science
                                                         institute name called ExcelR. You are very help AI BOT.
                                                         Your job is here to help students resolve their doubts regarding specific 
                                                         Data Science related topics. If the doubts is not relevant to the Data 
                                                         Science, you can politely ask the user to ask Data Science related doubt only 
                                                         and mentioned that you are a AI BOT and not trained with other than Data Science. 
                                                         Please don't answer other domain questions except data science."""},
                                                        {"role":"user",
                                                         "content":prompt}])
    return response.choices[0].message.content                                                         
```

```python
while True:
    user_input = input("User : ")
    if user_input.lower() in ["bye", "exit","close","quit", "see you"]:
        break
    response = chat_bot(user_input)
    print("**AI**---->>>>", response)
```

# Adding Memory to store the conversation history

```python
history = []
def chat_bot(prompt):
    response = client.chat.completions.create(model='gpt-4o-mini', 
                                              messages=[{"role":"system",
                                                         "content":"""You are a professional Education counsellor working with a Data Science
                                                         institute name called ExcelR. You are very help AI BOT.
                                                         Your job is here to help students resolve their doubts regarding specific 
                                                         Data Science related topics. If the doubts is not relevant to the Data 
                                                         Science, you can politely ask the user to ask Data Science related doubt only 
                                                         and mentioned that you are a AI BOT and not trained with other than Data Science. 
                                                         Please don't answer other domain questions except data science."""}] + 
                                              history + [{"role":"user", "content":prompt}])
    history.append({"role":"assistant", "content":response.choices[0].message.content })
    return response.choices[0].message.content                                                         
```

```python
while True:
    user_input = input("User : ")
    if user_input.lower() in ["bye", "exit","close","quit", "see you"]:
        break
    response = chat_bot(user_input)
    print("**AI**---->>>>", response)
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import requests

def image_from_url(image_url):
    img = Image.open(requests.get(image_url, stream=True).raw)
    return img

img = image_from_url(image_url)
plt.imshow(img)
```

```python
import io
original_img = image_from_url(image_url)
img_bytes = io.BytesIO()
original_img.save(img_bytes, format='PNG')
img_bytes = img_bytes.getvalue()
response = client.images.create_variation(image=img_bytes, n=3, size="1024x1024")
```

```python
for res in response.data:
    new_image_url = res.url
    new_img = image_from_url(new_image_url)
    plt.figure()
    plt.imshow(new_img)
```
